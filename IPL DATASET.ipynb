```python
# Find the most destructive death over batsman in the history of IPL
# Strike rate =(number of runs/number of balls played)*100
# Min batsman should have played 200 balls in over 16-20
```

```python
deliveries.head(3)
```

```python
temp_dataframe = deliveries[((deliveries["over"]>=16) & (deliveries["over"] <=20))]
temp_dataframe
```

```python
number_of_runs = temp_dataframe.groupby("batsman")["batsman_runs"].sum().sort_values()
number_of_runs
```

```python
tot_balls = temp_dataframe.groupby("batsman")["ball"].count().sort_values()
number_of_balls = tot_balls[(tot_balls >=200)]
number_of_balls
```

```python
strike_rate = (number_of_runs/number_of_balls)*100
strike_rate.sort_values(ascending=False).head(5)
```

```python
top_5_player = strike_rate.sort_values(ascending=False).head(5).index.tolist()
top_5_player
```

```python
details_of_top_5_players =deliveries[deliveries["batsman"].isin(top_5_player)]
details_of_top_5_players
```
